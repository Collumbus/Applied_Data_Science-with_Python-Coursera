<b>What is the approximation  for what the betweenness centrality of the nodes actually is?</b>

```python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('C', 'A'), ('A', 'E'), ('G', 'A'), 
                  ('A', 'N'), ('B', 'C'), ('D', 'B'), ('B', 'E'), 
                  ('C', 'D'), ('E', 'C'), ('D', 'E'), ('E', 'D'), 
                  ('F', 'G'), ('I', 'F'), ('J', 'F'), ('H', 'G'), 
                  ('I', 'G'), ('G', 'J'), ('I', 'H'), ('H', 'I'), 
                  ('I', 'J'), ('J', 'O'), ('O', 'J'), ('K', 'M'), 
                  ('K', 'L'), ('O', 'K'), ('O', 'L'), ('N', 'L'), 
                  ('L', 'M'), ('N', 'O')])
```

```python
# draw the graph using the default spring layout
plt.figure(figsize=(10,9))
nx.draw_networkx(G)
```

```python
# Paramter k says how many nodes you should use to compute the betweenness 
# centrality
btwnCent_approx = nx.betweenness_centrality(G,
normalized = True, endpoints = False, k=5)

sorted(btwnCent_approx.items(), key=lambda x: x[1], reverse=True)
```

<b>7. Given the network below, which value of alpha (damping parameter) listed below in the NetworkX function pagerank maximizes the PageRank of node D?</b>

```python
g = nx.DiGraph()
g.add_edges_from([('B', 'A'), ('A', 'C'), ('C', 'D'), ('D', 'C'), ('A', 'B')])

nx.draw_networkx(g)
```

```python
a_d = {}
for damp in [0.5, 0.8, 0.9, 0.95]:
    a_d[str(damp)] = nx.pagerank(G=g, alpha=damp)['D']

q7 = sorted(a_d.items(), key=lambda x: x[1], reverse=True)
q7[0][0]
```
