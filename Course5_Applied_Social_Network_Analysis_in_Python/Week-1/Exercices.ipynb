```python
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib inline

# Instantiate the graph
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([('A', 1),
                   ('A', 2),
                   ('A', 3),
                   ('A', 4),
                   ('A', 5),
                   ('B', 2),
                   ('C', 3),
                   ('C', 4),
                   ('C', 5),
                   ('D', 4),
                   ('E', 5)])

# draw the network G1
nx.draw_networkx(G1)
```

```python
from networkx.algorithms import bipartite
X = set(['A','B','C','D', 'E'])
P = bipartite.weighted_projected_graph(G1, X)
nx.draw_networkx(P)
```

```python
Y = set([1, 2, 3, 4, 5])
Q = bipartite.projected_graph(G1, Y)
nx.draw_networkx(Q)
```
