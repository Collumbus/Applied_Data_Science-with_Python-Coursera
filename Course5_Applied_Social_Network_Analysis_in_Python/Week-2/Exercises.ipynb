### #3

```python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('C', 'D'), ('C', 'E'), ('D', 'B'), ('D', 'E'), ('D', 'G'), ('G', 'E'), ('G', 'F')])
plt.figure(figsize=(10,9))
nx.draw_networkx(G)
```

```python
print(f'Radius: {nx.radius(G)}')
print(f'Diameter: {nx.diameter(G)}')
print(f'Periphery: {nx.periphery(G)}')
print(f'Center: {nx.center(G)}')
print(f'Eccentricity: {nx.eccentricity(G)}')
print(f'Breadth-First Search from node A: {nx.bfs_tree(G, "A").edges()}')
```

```python
G.remove_nodes_from('G')
plt.figure(figsize=(10,9))
nx.draw_networkx(G)
```

```python
print(f'Local clustering coefficient of node "I": {nx.clustering(G, "I")}')
print(f'Local clustering coefficient of node "J": {nx.clustering(G, "J")}')
print(f'Local clustering coefficient of node "K": {nx.clustering(G, "K")}')
print(f'Tue number of connected components is: {nx.number_connected_components(G)}')
```
