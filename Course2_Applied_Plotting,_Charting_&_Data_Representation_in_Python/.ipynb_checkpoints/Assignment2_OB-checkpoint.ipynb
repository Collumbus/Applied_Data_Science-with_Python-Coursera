```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import dateutil
import matplotlib.patches as mpatches

DATA_PATH = "data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv"

def day_of_year_no_leap(timestamp):
    if timestamp.is_leap_year and timestamp.month >= 3:
        return timestamp.dayofyear-1
    return timestamp.dayofyear
        
def load_station_data():
    df = pd.read_csv(DATA_PATH)
    
    df['Date'] = df['Date'].apply(dateutil.parser.parse)
    df['Day_of_Year'] = df['Date'].apply(day_of_year_no_leap)

    # Remove leap-days
    df = df[df['Date'].apply(lambda d: not(d.day==29 and d.month==2))]
    
    # Split into 2009-2014 and 2015
    df_2015 = df[df['Date'].apply(lambda d: d.year==2015)]
    df_prev = df[df['Date'].apply(lambda d: d.year<2015)]
    return df_prev, df_2015
    
df_prev, df_2015 = load_station_data()
df_prev.info()
df_2015.info()
```

```python
import numpy as np
t_prev = pd.pivot_table(df_prev, 
                        index=['Day_of_Year'], 
                        columns=['Element'], 
                        values=['Data_Value'], 
                        aggfunc=[min, max, np.mean])
t_2015 = pd.pivot_table(df_2015, 
                        index=['Day_of_Year'], 
                        columns=['Element'], 
                        values=['Data_Value'], 
                        aggfunc=[min, max, np.mean])

temp_mins = t_prev[('min', 'Data_Value', 'TMIN')]/10.
temp_maxs = t_prev[('max', 'Data_Value', 'TMAX')]/10.

tmp_t2015_mins = t_2015[('min', 'Data_Value', 'TMIN')]/10.
tmp_t2015_maxs = t_2015[('max', 'Data_Value', 'TMAX')]/10.

t2015_mins = t_2015[tmp_t2015_mins < temp_mins]
t2015_maxs = t_2015[tmp_t2015_maxs > temp_maxs]

```

```python
for y in range(-30,41,10):
    plt.axhline(y=y,
                ls='-',
                color='black',
                linewidth=0.5,
                alpha=0.15)

plt.plot(range(1,366), 
         temp_mins,
         label="Temperature range 2014-2015"
         "-", 
         linewidth=0.5,
         color="gray",
         alpha=0.25)#color="#0000ff")

plt.plot(range(1,366), 
         temp_maxs, 
         "-", 
         linewidth=0.5,
         color="gray",
         alpha=0.25)#color="#0000ff")

plt.gca().fill_between(range(1,366), 
                       temp_mins, temp_maxs, 
                       facecolor='gray', 
                       alpha=0.25)

max_points = plt.scatter(t2015_maxs.index,
            t2015_maxs[('max', 'Data_Value', 'TMAX')]/10.,
            marker="2",
            color="#ff0000",
            label="2015 measurement above past nine year record")

min_points = plt.scatter(t2015_mins.index,
            t2015_mins[('min', 'Data_Value', 'TMIN')]/10.,
            marker="1",
            color="#0000ff",
            label="2015 measurement below past nine year record")

plt.gca().set_xlim([0, 365])

#plt.xlabel('Month')
#plt.ylabel('Degrees celsius')

# x-axis ticks
days_per_month = [
    ("January", 31),
    ("February", 28),
    ("March", 31),
    ("April", 30),
    ("May", 31),
    ("June", 30),
    ("July", 31),
    ("August", 31),
    ("September", 30),
    ("October", 31),
    ("November", 30),
    ("December", 31)
    ]

days_total = 0
month_ticks = []
for month, days in days_per_month:
    month_ticks.append((month, days_total+days/2.0))
    days_total += days
    
month_names, days = list(zip(*month_ticks))

plt.gca().set_xticks(days)
plt.gca().xaxis.set_ticks_position('none')
plt.gca().set_xticklabels(month_names)

#x = plt.gca().xaxis
# rotate the tick labels for the x axis
#for item in x.get_ticklabels():
#    item.set_rotation(45)

# Figure size
plt.gcf().set_size_inches(18.5, 10.5)
"""
x_running = 0
for _, days in days_per_month[:-1]:
    x_running += days
    plt.axvline(x=x_running,
                ls='-',
                color='gray',
                alpha=0.15)
"""

#plt.gca().axis('off')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.gca().set_yticks(range(-30,41,10))
plt.gca().set_yticklabels(
    ["${0} \degree$C".format(t) for t in range(-30,41,10)])
plt.gca().yaxis.set_ticks_position('none')

#plt.annotate('Range of tempertures 2004-2015', xy=(75, 13), xytext=(30, 25),
#            arrowprops=dict(facecolor='black', shrink=0.1, width=1.0, headwidth=9.0),horizontalalignment='center'
#            )

import matplotlib.lines as mlines
import matplotlib.patches as mpatches

red_patch = mpatches.Patch(color='gray', alpha=0.25, label='Range of temperatures measured from 2005-2014')

plt.legend(handles=[red_patch, max_points, min_points])
plt.title("Days with temperatures outside 2005-2014 record highs and lows in the Ann Arbor area")

plt.show()
```
